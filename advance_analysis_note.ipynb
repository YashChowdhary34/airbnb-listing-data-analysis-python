```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline
```

**4. Data Cleaning**

```python
#Check for missing values
data.isnull().sum()
#Drop the missing values
data.dropna(inplace=True)
#Fill the missing values with 0
data.fillna(0, inplace=True)
```

```python
#Check for duplicates
data.duplicated().sum()
#Get the duplicates
data[data.duplicated()]
#Drop the duplicates
data.drop_duplicates(inplace=True)
```

```python
#List the data types of the columns
data.dtypes
#Change the data type of a column
data['id'] = data['id'].astype('object')
data['host_id'] = data['host_id'].astype('object')
```

### EDA
**5. Data Analysis**

**Univariate Analysis**


```python
#identifying outliers in price
plt.figure(figsize=(10, 6))
df = data[data['price'] < 1500]
sns.boxplot(data=data, x='price')
```

```python
#Price distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='price', bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
#Room availability distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='availability_365')
plt.title('Availability Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean().sort_values(ascending=False)
```

**Feature Engineering**


```python
#Creating new column with price per bed

df.loc[:, 'price_per_bed'] = df['price'] / df['beds'].replace(0, np.nan) 
df['price_per_bed'].fillna(0, inplace=True) 
df.head()
```

```python
df.groupby(by='neighbourhood_group')['price_per_bed'].mean().sort_values(ascending=False)
```

**BI Variable Analysis**
One variable dependency in another variable

```python
df.columns
```

```python
#price_per_bed dependency on neighbourhood_group
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='neighbourhood_group', y='price_per_bed', hue='room_type')
plt.title('Price per bed dependency on neighbourhood_group')
```

```python
#number of reviews dependency on price

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='price', y='number_of_reviews', hue='neighbourhood_group')
plt.title('Number of reviews dependency on price')
```

```python
#pairplot of the data according to neighbourhood_group
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365'], hue='neighbourhood_group')
plt.title('Pairplot of the data according to neighbourhood_group')
```

```python
#geographical distribution of the all the listings

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical distribution of the all the listings')
```

```python
#heat map for correlation of one variable with others for all the numerical columns

corr = df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(data=corr, annot=True)
```
